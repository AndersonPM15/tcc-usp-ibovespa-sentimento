```python
import pandas as pd
import numpy as np
from pathlib import Path
from src.io import paths
from src.config.constants import START_DATE, END_DATE

DATA_PATHS = paths.get_data_paths()
RAW = DATA_PATHS['data_raw']
PROCESSED = DATA_PATHS['data_processed']

print(f"Período planejado: {START_DATE} a {END_DATE}")
print(f"Dias teóricos: {(END_DATE - START_DATE).days + 1}")
```

## 1. ARQUIVOS BRUTOS (data_raw)

```python
# IBOVESPA BRUTO
print("="*80)
print("IBOVESPA.CSV (bruto)")
print("="*80)
ibov_raw = pd.read_csv(RAW / 'ibovespa.csv', parse_dates=['date'])
print(f"Shape: {ibov_raw.shape}")
print(f"Colunas: {list(ibov_raw.columns)}")
print(f"Min date: {ibov_raw['date'].min()}")
print(f"Max date: {ibov_raw['date'].max()}")
print(f"Dias distintos: {ibov_raw['date'].nunique()}")
print(f"\nPrimeiras linhas:")
print(ibov_raw.head(3))
print(f"\nÚltimas linhas:")
print(ibov_raw.tail(3))
```

```python
# NOTÍCIAS MULTISOURCE BRUTAS
print("="*80)
print("NEWS_MULTISOURCE.PARQUET (bruto)")
print("="*80)
news_raw = pd.read_parquet(RAW / 'news_multisource.parquet')
if 'date' in news_raw.columns:
    news_raw['date'] = pd.to_datetime(news_raw['date'], errors='coerce')
elif 'published_at' in news_raw.columns:
    news_raw['date'] = pd.to_datetime(news_raw['published_at'], errors='coerce')
print(f"Shape: {news_raw.shape}")
print(f"Colunas: {list(news_raw.columns)}")
print(f"Min date: {news_raw['date'].min()}")
print(f"Max date: {news_raw['date'].max()}")
print(f"Dias distintos: {news_raw['date'].dt.date.nunique()}")
print(f"Fontes: {news_raw['source'].value_counts().to_dict() if 'source' in news_raw.columns else 'N/A'}")
print(f"\nPrimeiras linhas:")
print(news_raw.head(2))
```

## 2. ARQUIVOS PROCESSADOS (data_processed)

```python
# IBOVESPA LIMPO
print("="*80)
print("IBOV_CLEAN.CSV (processado)")
print("="*80)
ibov_clean = pd.read_csv(PROCESSED / 'ibov_clean.csv')
# Tentar identificar coluna de data
date_col = None
for col in ['day', 'date', 'Date', 'DATA']:
    if col in ibov_clean.columns:
        date_col = col
        break
if date_col:
    ibov_clean[date_col] = pd.to_datetime(ibov_clean[date_col])
    print(f"Shape: {ibov_clean.shape}")
    print(f"Colunas: {list(ibov_clean.columns)}")
    print(f"Min date: {ibov_clean[date_col].min()}")
    print(f"Max date: {ibov_clean[date_col].max()}")
    print(f"Dias distintos: {ibov_clean[date_col].nunique()}")
else:
    print(f"Shape: {ibov_clean.shape}")
    print(f"Colunas: {list(ibov_clean.columns)}")
    print("ERRO: Coluna de data não encontrada!")
print(f"\nPrimeiras linhas:")
print(ibov_clean.head(3))
print(f"\nÚltimas linhas:")
print(ibov_clean.tail(3))
```

```python
# NOTÍCIAS LIMPAS
print("="*80)
print("NEWS_CLEAN.PARQUET (processado)")
print("="*80)
news_clean = pd.read_parquet(PROCESSED / 'news_clean.parquet')
date_col = None
for col in ['day', 'date', 'Date']:
    if col in news_clean.columns:
        date_col = col
        break
if date_col:
    news_clean[date_col] = pd.to_datetime(news_clean[date_col])
    print(f"Shape: {news_clean.shape}")
    print(f"Colunas: {list(news_clean.columns)}")
    print(f"Min date: {news_clean[date_col].min()}")
    print(f"Max date: {news_clean[date_col].max()}")
    print(f"Dias distintos: {news_clean[date_col].dt.date.nunique()}")
else:
    print(f"Shape: {news_clean.shape}")
    print("ERRO: Coluna de data não encontrada!")
print(f"\nPrimeiras linhas:")
print(news_clean.head(2))
```

```python
# MATRIZ TF-IDF DIÁRIA
print("="*80)
print("TFIDF_DAILY_INDEX.CSV (processado - índice da matriz)")
print("="*80)
tfidf_index = pd.read_csv(PROCESSED / 'tfidf_daily_index.csv', parse_dates=['day'])
print(f"Shape: {tfidf_index.shape}")
print(f"Colunas: {list(tfidf_index.columns)}")
print(f"Min date: {tfidf_index['day'].min()}")
print(f"Max date: {tfidf_index['day'].max()}")
print(f"Dias distintos: {tfidf_index['day'].nunique()}")
print(f"\nPrimeiras linhas:")
print(tfidf_index.head(3))
print(f"\nÚltimas linhas:")
print(tfidf_index.tail(3))
```

```python
# LABELS Y DIÁRIOS
print("="*80)
print("LABELS_Y_DAILY.CSV (processado - targets)")
print("="*80)
y_labels = pd.read_csv(PROCESSED / 'labels_y_daily.csv', parse_dates=['day'])
print(f"Shape: {y_labels.shape}")
print(f"Colunas: {list(y_labels.columns)}")
print(f"Min date: {y_labels['day'].min()}")
print(f"Max date: {y_labels['day'].max()}")
print(f"Dias distintos: {y_labels['day'].nunique()}")
print(f"\nDistribuição do target:")
if 'target_1d' in y_labels.columns:
    print(y_labels['target_1d'].value_counts())
print(f"\nPrimeiras linhas:")
print(y_labels.head(3))
```

```python
# PREDIÇÕES OOF
print("="*80)
print("16_OOF_PREDICTIONS.CSV (predições do modelo)")
print("="*80)
oof = pd.read_csv(PROCESSED / '16_oof_predictions.csv', parse_dates=['day'])
print(f"Shape: {oof.shape}")
print(f"Colunas: {list(oof.columns)}")
print(f"Min date: {oof['day'].min()}")
print(f"Max date: {oof['day'].max()}")
print(f"Dias distintos: {oof['day'].nunique()}")
print(f"\nModelos: {oof['model'].unique() if 'model' in oof.columns else 'N/A'}")
print(f"\nPrimeiras linhas:")
print(oof.head(3))
```

```python
# BACKTEST RESULTS
print("="*80)
print("18_BACKTEST_RESULTS.CSV (resultados de backtesting)")
print("="*80)
backtest = pd.read_csv(PROCESSED / '18_backtest_results.csv')
print(f"Shape: {backtest.shape}")
print(f"Colunas: {list(backtest.columns)}")
print(f"\nModelos e estratégias:")
print(backtest.groupby(['model', 'strategy']).size())
print(f"\nPrimeiras linhas:")
print(backtest.head(5))
```

## 3. TABELA RESUMO DA AUDITORIA

```python
# Criar tabela resumo
audit_data = []

# Ibovespa bruto
audit_data.append({
    'Arquivo': 'ibovespa.csv (RAW)',
    'Min Date': ibov_raw['date'].min(),
    'Max Date': ibov_raw['date'].max(),
    '# Linhas': len(ibov_raw),
    '# Dias': ibov_raw['date'].nunique()
})

# Notícias brutas
audit_data.append({
    'Arquivo': 'news_multisource.parquet (RAW)',
    'Min Date': news_raw['date'].min(),
    'Max Date': news_raw['date'].max(),
    '# Linhas': len(news_raw),
    '# Dias': news_raw['date'].dt.date.nunique()
})

# Ibovespa limpo
if date_col:
    audit_data.append({
        'Arquivo': 'ibov_clean.csv (PROCESSED)',
        'Min Date': ibov_clean[date_col].min(),
        'Max Date': ibov_clean[date_col].max(),
        '# Linhas': len(ibov_clean),
        '# Dias': ibov_clean[date_col].nunique()
    })

# TF-IDF index
audit_data.append({
    'Arquivo': 'tfidf_daily_index.csv (PROCESSED)',
    'Min Date': tfidf_index['day'].min(),
    'Max Date': tfidf_index['day'].max(),
    '# Linhas': len(tfidf_index),
    '# Dias': tfidf_index['day'].nunique()
})

# Labels Y
audit_data.append({
    'Arquivo': 'labels_y_daily.csv (PROCESSED)',
    'Min Date': y_labels['day'].min(),
    'Max Date': y_labels['day'].max(),
    '# Linhas': len(y_labels),
    '# Dias': y_labels['day'].nunique()
})

# OOF Predictions
audit_data.append({
    'Arquivo': '16_oof_predictions.csv (PROCESSED)',
    'Min Date': oof['day'].min(),
    'Max Date': oof['day'].max(),
    '# Linhas': len(oof),
    '# Dias': oof['day'].nunique()
})

audit_df = pd.DataFrame(audit_data)
print("\n" + "="*100)
print("TABELA RESUMO DA AUDITORIA")
print("="*100)
print(audit_df.to_string(index=False))

# Salvar para referência
audit_df.to_csv(PROCESSED / 'AUDIT_pipeline_summary.csv', index=False)
print("\n✅ Tabela salva em: data_processed/AUDIT_pipeline_summary.csv")
```

## 4. DIAGNÓSTICO: ONDE ESTÁ O ESTRANGULAMENTO?

```python
print("\n" + "="*100)
print("DIAGNÓSTICO DO ESTRANGULAMENTO")
print("="*100)

print(f"\n1. IBOVESPA:")
print(f"   - Bruto: {ibov_raw['date'].nunique()} dias ({ibov_raw['date'].min()} a {ibov_raw['date'].max()})")
if date_col:
    print(f"   - Limpo: {ibov_clean[date_col].nunique()} dias ({ibov_clean[date_col].min()} a {ibov_clean[date_col].max()})")
    perda_ibov = ibov_raw['date'].nunique() - ibov_clean[date_col].nunique()
    print(f"   - ⚠️ PERDA: {perda_ibov} dias ({perda_ibov/ibov_raw['date'].nunique()*100:.1f}%)")

print(f"\n2. NOTÍCIAS:")
print(f"   - Brutas: {news_raw['date'].dt.date.nunique()} dias com notícias")
print(f"   - TF-IDF: {tfidf_index['day'].nunique()} dias com features")
perda_news = news_raw['date'].dt.date.nunique() - tfidf_index['day'].nunique()
print(f"   - ⚠️ PERDA: {perda_news} dias ({perda_news/news_raw['date'].dt.date.nunique()*100:.1f}%)")

print(f"\n3. FEATURES + LABELS (merge):")
print(f"   - TF-IDF index: {tfidf_index['day'].nunique()} dias")
print(f"   - Labels Y: {y_labels['day'].nunique()} dias")
print(f"   - OOF predictions: {oof['day'].nunique()} dias")
perda_merge = tfidf_index['day'].nunique() - oof['day'].nunique()
print(f"   - ⚠️ PERDA NO MERGE: {perda_merge} dias ({perda_merge/tfidf_index['day'].nunique()*100:.1f}%)")

print(f"\n4. DASHBOARD:")
print(f"   - Mostra apenas {ibov_clean.shape[0]} linhas de Ibovespa (esperado: ~{ibov_raw['date'].nunique()})")
print(f"   - Mostra apenas {oof['day'].nunique()} dias de sentimento (esperado: ~{tfidf_index['day'].nunique()})")

print(f"\n" + "="*100)
print("CONCLUSÃO:")
print("="*100)
if ibov_clean.shape[0] < 50:
    print("🚨 PROBLEMA CRÍTICO: ibov_clean.csv tem apenas 20 linhas!")
    print("   Isso sugere que o arquivo foi sobrescrito com dados de teste/exemplo.")
if oof['day'].nunique() < 50:
    print("🚨 PROBLEMA CRÍTICO: Predições OOF têm apenas 12 dias!")
    print("   Isso sugere que os modelos foram treinados apenas em uma amostra mínima.")
```
